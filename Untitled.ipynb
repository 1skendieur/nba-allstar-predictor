To build a strong predictor of NBA All-Star participation, we trained and evaluated multiple machine learning models using the preprocessed dataset.

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)
X_test = scaler.transform(X_test)
```

```python
early_stop = EarlyStopping(
    monitor='val_loss',         # what to monitor
    patience=5,                 # how many epochs to wait without improvement
    restore_best_weights=True  # use the best model seen during training
)

model = keras.Sequential([
    keras.Input(shape=(X_train.shape[1],)),
    layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    layers.Dropout(0.4),
    layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    layers.Dropout(0.3),
    layers.Dense(64, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.0005),
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=100,
    batch_size=32,
    callbacks=[early_stop],
    verbose=1
)
```

```python
y_val_pred_nn = (model.predict(X_val) > 0.5).astype(int)
nn_val_acc = accuracy_score(y_val, y_val_pred_nn)
```

```python
model_results = [
    ("Neural Network", nn_val_acc)
]

# Logistic Regression
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)
logreg_val_acc = accuracy_score(y_val, logreg.predict(X_val))
model_results.append(("Logistic Regression", logreg_val_acc))

# Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
rf_val_acc = accuracy_score(y_val, rf.predict(X_val))
model_results.append(("Random Forest", rf_val_acc))

# XGBoost
xgb = XGBClassifier(eval_metric='logloss')
xgb.fit(X_train, y_train)
xgb_val_acc = accuracy_score(y_val, xgb.predict(X_val))
model_results.append(("XGBoost", xgb_val_acc))

# SVM
svm = SVC(probability=True)
svm.fit(X_train, y_train)
svm_val_acc = accuracy_score(y_val, svm.predict(X_val))
model_results.append(("SVM", svm_val_acc))
```

```python
print("Accuracy Comparison:")
for name, acc in model_results:
    print(f"{name:20s}: {acc * 100:.2f}%")
```

```python
names = [x[0] for x in model_results]
scores = [x[1] * 100 for x in model_results]

plt.figure(figsize=(10, 6))
bars = plt.barh(names, scores, color='skyblue', edgecolor='black')

# Annotate bars with values
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.5, bar.get_y() + bar.get_height() / 2,
             f'{width:.2f}%', va='center', fontsize=10)

plt.xlabel('Validation Accuracy (%)')
plt.xlim(min(scores) - 2, 100)  # Add margin to better see differences
plt.title('Model Comparison')
plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
best_model = xgb
y_test_pred = best_model.predict(X_test)
print("Test Accuracy:", accuracy_score(y_test, y_test_pred))
print(classification_report(y_test, y_test_pred))
```

### âœ… Post-Modeling Steps (Before API Deployment)
After picking the best-performing model (XGBoost), we did some important steps after the modelling to understand how well it worked and get it ready for real-life use.

```python
y_probs = best_model.predict_proba(X_test)[:, 1]

threshold = 0.4
y_pred_threshold = (y_probs >= threshold).astype(int)

print("Threshold:", threshold)
print(confusion_matrix(y_test, y_pred_threshold))
print(classification_report(y_test, y_pred_threshold, target_names=["Not All-Star", "All-Star"]))
```

```python
feature_columns = X.columns

def predict_all_star_chance(player_name, df, model, scaler, feature_columns):
    # 1. Find player row
    row = df[df["Player.x"] == player_name]
    if row.empty:
        return f"Player '{player_name}' not found in dataset."

    # 2. Drop label & preprocess
    X_player = row[feature_columns].copy()
    X_player = pd.get_dummies(X_player)

    # 3. Align columns
    X_player = X_player.reindex(columns=scaler.feature_names_in_, fill_value=0)

    # 4. Scale
    X_player_scaled = scaler.transform(X_player)

    # 5. Predict probability
    prob = model.predict_proba(X_player_scaled)[0][1]  # class 1 (All-Star)
    return f"{player_name}'s chance of being an All-Star: {prob * 100:.2f}%"

print(predict_all_star_chance("LeBron James", df, best_model, scaler, feature_columns))
print(predict_all_star_chance("Kyrie Irving", df, best_model, scaler, feature_columns))
print(predict_all_star_chance("Draymond Green", df, best_model, scaler, feature_columns))
```

### ðŸ“Ÿ Using the REST API to deploy the model
To make the model useful in the real world, we used it as a RESTful API with FastAPI. This allows clients or other services to send player statistics and receive a prediction of how many players will be selected for the All-Stars.

## Creating the Prediction API
We created an API.py file using FastAPI. It uses Pydantic to validate data strictly and provides a /predict endpoint.

It can handle player stats in JSON format.

Use the same scaler to measure the input as you did in training.

This model uses something called the XGBoost algorithm to predict which players will be in the All-Star team and how likely they are of being picked.

It will give you a JSON response with the class and confidence.

API.py code:
from fastapi import FastAPI, HTTPException
from pydantic import BaseModel, Field
import joblib
import pandas as pd
import numpy as np

model = joblib.load("best_model_xgb.pkl")
scaler = joblib.load("scaler.pkl")

class PlayerFeatures(BaseModel):
    Pos1: int
    Age: float
    Tm: int
    G: float
    GS: float
    MP: float
    FG: float
    FGA: float
    FG_dot: float = Field(alias="FG.")
    X3P: float
    X3PA: float
    X3P_dot: float = Field(alias="X3P.")
    X2P: float
    X2PA: float
    X2P_dot: float = Field(alias="X2P.")
    eFG_dot: float = Field(alias="eFG.")
    FT: float
    FTA: float
    FT_dot: float = Field(alias="FT.")
    ORB: float
    DRB: float
    TRB: float
    AST: float
    STL: float
    BLK: float
    TOV: float
    PF: float
    PTS: float
    Salary: float
    mean_views: float
    Conference: int
    Role: int
    Fvot: float
    FRank: float
    Pvot: float
    PRank: float
    Mvot: float
    MRank: float
    Score: float

app = FastAPI()

@app.post("/predict/")
def predict(data: PlayerFeatures):
    try:
        input_df = pd.DataFrame([data.dict(by_alias=True)])
        scaled_input = scaler.transform(input_df)
        prob = model.predict_proba(scaled_input)[0][1]
        prediction = "All-Star" if prob >= 0.5 else "Not All-Star"

        return {
            "all_star_probability": float(round(prob * 100, 2)),  # ÐŸÑ€ÐµÐ¾Ð±Ñ€Ð°Ð·ÑƒÐµÐ¼ Ð² Ð¾Ð±Ñ‹Ñ‡Ð½Ñ‹Ð¹ float
            "prediction": prediction
        }

    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))
